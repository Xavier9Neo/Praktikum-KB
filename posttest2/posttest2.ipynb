3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.

tr = total record
ta = total attribute
na = nama attribute

```python
tr = stroke.size
ta = len(stroke.columns)
na = stroke.columns
jumlah = len(stroke.select_dtypes('int64').columns)

data = {
    "Jumlah Record" : tr,
    "Jumlah Attribute" : ta,
    "Nama Attribute" : na,
    "Jumlah Attribute tipe angka" : jumlah
}

tabel = PrettyTable()
for i in data.keys():
    tabel.add_column(i, [])
tabel.add_row([data.get(i, "")for i in data.keys()])
print(tabel)
```
